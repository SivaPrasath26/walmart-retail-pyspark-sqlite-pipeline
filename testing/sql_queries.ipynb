```python
# Cell 1: Establish connection
import sqlite3
import pandas as pd

conn = sqlite3.connect("data/curated/retail_pipeline.db")
```

```python
# Query all columns
query = "SELECT * FROM sales_curated;"
df = pd.read_sql_query(query, conn)
df
```

```python
#Total Sales
query1 = "SELECT store_id, SUM(weekly_sales) AS total_sales FROM sales_curated GROUP BY store_id ORDER BY total_sales DESC LIMIT 10;"
df1 = pd.read_sql_query(query1, conn)
df1
```

```python
# Average Sales
query2 = "SELECT store_type, AVG(weekly_sales) AS avg_sales FROM sales_curated GROUP BY store_type;"
df2 = pd.read_sql_query(query2, conn)
df2
```
